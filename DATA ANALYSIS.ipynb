Create freq_table() function to find most common geners

To find the frequency distribution of a specific column in a dataset, which in this case is the genre column.

```python
def freq_table(dataset, index):
    table = {}
    total = 0
    for app_row in dataset:
        total += 1
        item = app_row[index]
        if item in table:
            table[item] += 1
        else:
            table[item] = 1
    freq_percentage = {}
    for key in table:
        freq_percentage[key] = (table[key] / total) * 100
    return freq_percentage   
# Call the function with the correct column name
    freq_table(df, 'prime_genre')
```


Inspecting linear correlation of Number of languages and Rating Count (Total)

```python
import plotly.express as px
import statsmodels.api as sm

def visualizer(x, y, title, x_label, y_label, rotate_x_labels, angle):
    X = sm.add_constant(x)
    model = sm.OLS(y, X).fit()
    predictions = model.predict(X)

    fig = px.scatter(x=x, y=y, title=title, labels={x_label: x_label, y_label: y_label})
    fig.add_traces(px.line(x=x, y=predictions).data)

    if rotate_x_labels:
        fig.update_layout(xaxis_tickangle=angle)

    fig.show()
```

```python
visualizer(df["lang.num"], df.rating_count_tot, 
          "CORRELATION OF NUMBER OF LANGUAGES AND RATING COUNT", "NUMBER OF LANGAUGES",
          "RATING COUNT (TOTAL)", False, 0)
```
